```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression


np.random.seed(42)
X = np.random.uniform(0, 50, 100)


noise = np.random.normal(0, 10, 100)
Y = 3*X + 5 + noise

X_reshaped = X.reshape(-1,1)

slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)


X_sort = np.sort(X)


Y_pred = slope * X_sort + intercept 
# --- FIX END ---

print("Method 1 (scipy)")
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r_value)
print("P-value:", p_value)
print("Standard Error:", std_err)

plt.figure(figsize=(10,6))
plt.scatter(X, Y, label='Data Points')
plt.plot(X_sort, Y_pred, color='red', label='Regression Line')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 8, 8, 10])

# Step 1: Calculate mean
x_mean = np.mean(x)
y_mean = np.mean(y)

# Step 2: Calculate slope (m) and intercept (c)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)

m = numerator / denominator
c = y_mean - m * x_mean

print("Slope (m):", m)
print("Intercept (c):", c)

# Step 3: Predicted y values
y_pred = m * x + c

# Step 4: Calculate MSE and RSS
mse = np.mean((y - y_pred)**2)
rss = np.sum((y - y_pred)**2)

print("MSE:", mse)
print("RSS:", rss)

# Step 5: Predict for new x value
x_new = 6
y_new = m * x_new + c
print("Predicted y for x = 6 :", y_new)
plt.scatter(x, y, label="Original Data")
plt.plot(x, y_pred, color='red', label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

```
