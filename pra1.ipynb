```python
#list
#list is mutable i can change index

data = [1, 2, 43, 4]
print(data)

print(data[0])
data[0] = 44
print(data[0])
print(data[-1])
data[0] = 4
print(data)

numbers = [0, 1, 2, 3, 4, 5]
print(numbers[1:2])
print(numbers[1:5:3])

fruits = ["apple", "banana"]
fruits.append("cherry")
print(fruits)

fruits.insert(1, "mango")
print(fruits)

fruits.append("ss")
fruits.pop()
print(fruits)

num = [1, 3, 4, 5, 6, 4]
num.sort()
print(num)
```

```python
#tuple is unmutabel
data = (1.3, 4.2, 3.4, 1.3)
print(data[0])
print(data[2])
print(data)

data2 = (3, 4, 4, 5, 5, 5)
print(data2[2:5])
print(data2[1:5])

original = (1, 2, 3, 3, 9, 5, 5)
temp_list = original + (4,)
print(temp_list)

temp_list = list(temp_list)
print(temp_list)
temp_list.remove(2)
print(temp_list)

temp_list.sort()
print(temp_list)

temp_list = tuple(temp_list)
print(temp_list)



```

```python
#set
#set is also unmutable
s = {1, 2, 3, 4, 5}
print(s)

s1 = list(s)
print(s1)
print(s1[1:3])
print(s1[2:4:2])

s1.append(22)
print(s1)

s1.insert(2, 22)
print(s1)

s1.remove(2)
print(s1)
s1.pop()
print(s1)
s1.sort()
print(s1)
```

```python
#dictionary
dict = {
    "name": "jalaj",
    "language": "python",
    "uni": "sc"
}

print(dict)
print(type(dict))
print(dict["name"])

dict["id"] = "3321313"
print(dict)

dict["studying year"] = 2020
print(dict)

del dict["name"]
print(dict)

scores = {"jalaj": 88, "Al": 95, "B": 75}
sorted_keys = sorted(scores)
print(sorted_keys)

sorted_dict = {k: scores[k] for k in sorted_keys}
print(sorted_dict)
```
