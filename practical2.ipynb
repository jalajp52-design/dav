```python
#tuple is unmutable so i want change type firast
original = (1, 2, 3,3,9,5,5)
temp_list = original + (4,) 
print(temp_list)

temp_list = list(temp_list)

print(f"{temp_list}type of this {type(temp_list)}")

temp_list.remove(2) 
print(f"{temp_list}type of this {type(temp_list)}")

temp_list.sort()
print(f"{temp_list}type of this {type(temp_list)}")
temp_list = tuple(temp_list)
print(f"{temp_list}type of this {type(temp_list)}")

```

```python
#set
#set is also unmutable
s = {1, 2, 3, 4, 5}
print(s)

s1 = list(s)
print(type(s1))
print(s1[0])

#indexing in c
print(s1[1:3])
print(s1[2:4:2])

s1.append(22)
print(s1)

s1.insert(2, 22)   
print(s1)

s1.remove(2)
print(s1)

s1.pop()
print(s1)

s1.sort()   
print(s1)
```

```python
#dictionaries
dict={
"name":"jalaj",
"language":"python",
"uni":"sc"

}
print(dict)
print(type(dict))
print(dict["name"])
# ADDING a new key-value pair
dict["id"] = "3321313"
print(dict) 
# Output: {'brand': 'Ford', 'model': 'Mustang', 'year': 1964, 'color': 'red'}

# UPDATING an existing key
dict["studying year"] = 2020
print(dict) # Output: 2020
del dict["name"]
print(dict)
```

```python
scores = {"Charlie": 88, "Alice": 95, "Bob": 75}

# Returns a list of keys sorted alphabetically
sorted_keys = sorted(scores)
print(sorted_keys)  # Output: ['Alice', 'Bob', 'Charlie']

# Create a new sorted dictionary from this
sorted_dict = {k: scores[k] for k in sorted_keys}
print(sorted_dict)  # Output: {'Alice': 95, 'Bob': 75, 'Charlie': 88}
```
