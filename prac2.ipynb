```python
import numpy as np 
import random  
# Create a 1D array with 12 elements 
arr = np.arange(12)  
print("Original:", arr) 

# Reshape into a 2D matrix (3 rows, 4 columns) 
matrix = arr.reshape(3, 4) 
print("\nReshaped (3x4):\n", matrix) 

# Check attributes 
print(f"\nShape: {matrix.shape}") # (3, 4) 
print(f"Dimensions: {matrix.ndim}") # 2 
print(f"Data Type: {matrix.dtype}") # int32 or int64 
# The element at index [row=2, col=3] 
element = matrix[2, 3] 
print(f"Element at [2, 3]: {element}") 
```

```python
import numpy as np 
arr = np.array([5, 2, 9, 1, 7]) 
sorted_arr = np.sort(arr) 
print("Originalarray:", arr) 
print("Sortedarray:", sorted_arr) 
import numpy as np 
arr = np.array([ 
    [3, 1, 2], 
    [6, 4, 5] 
 
]) 
row_sort = np.sort(arr, axis=1) 
col_sort = np.sort(arr, axis=0) 
print("Originalarray:\n", arr) 
print("Row-wisesorted:\n", row_sort) 
print("Column-wisesorted:\n", col_sort)
```
