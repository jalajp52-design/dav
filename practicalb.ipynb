```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 8, 8, 10])

x_mean = np.mean(x)
y_mean = np.mean(y)

print(x_mean)
print(y_mean)
#sum x-xmean*y-ymean/x-xmean*2
numratator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)

m = numratator / denominator
c = y_mean - m * x_mean

print("Slope ", m)
print("Intercept ", c)
#y=mx+c
y_h=m*x_mean+c
y_h = m * x + c

mse = np.mean((y - y_h)** 2)
rss = np.sum((y - y_h) **2)

plt.scatter(x, y)
plt.plot(x, y_h)

plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
