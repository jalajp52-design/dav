```python
# -*- coding: utf-8 -*- 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 

# 1. Define the data
x = [5,7,8,7,2,17,2,9,4,11,12,9,6] 
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] 

# 2. Prepare Training Data
# Convert list to numpy array
Train_x = np.array(x) 
print("Train_x Before Reshaping: ", Train_x) 

# Reshape X for sklearn (converts 1D array to 2D column vector)
Train_x = Train_x.reshape(-1, 1) 
print("Train_x After Reshaping:\n", Train_x) 

# 3. Train the Model
Model = LinearRegression() 
Model.fit(Train_x, y) 

# 4. Prepare Testing Data (Using same data for demonstration)
Test_x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) 
Test_y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) 

# Reshape Test X
Test_x = Test_x.reshape(-1, 1) 

# 5. Make Predictions
y_pred = Model.predict(Test_x) 

# 6. Output Results
print("\n--- Results ---")
print("Actual Values (Test_y):", Test_y)
print("Predicted Values (y_pred):", y_pred)

# Calculate R^2 Score
accuracy = Model.score(Test_x, Test_y)

# Print Score using .format() for compatibility with older Python versions
print("\nModel R^2 Score (Accuracy): {:.4f}".format(accuracy))

# 7. Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(Test_x,Test_y, color='black', label='Actual Data')
plt.plot(Test_x, y_pred, color='blue', linewidth=3, label='Regression Line')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.title('Linear Regression Fit')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
